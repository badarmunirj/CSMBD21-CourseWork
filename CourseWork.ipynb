<h1> Task 1 <h1>

```python
import numpy as np
import pandas as pd
import json
#mapper
passenger='AComp_Passenger_data_no_error_DateTime.csv'
results = []

#reading line by line from the file
for elements in open(passenger):
    # Split the line into records - list of columns 
    element=elements.strip()
    element=elements.split(',')
    #saving Airport Code & Flight number in a variable
    new=('%s,%s'%(element[2],element[1]))
    #appending the values into a list
    results.append(new)
    
#mapper output
results
```

```python
#sorting
words=list()

#input
for line in results:
    #taking the first word
    word=line.strip().split()
    # Append to list words
    words.append(word)
    
# Sort list
words.sort()
#printing the output 
for line in words:
    print(line)
```

```python
#reducer
flight_dict={}
#reading line by line from the output of mapper
for line in words:
    for line1 in line:
        x,y=line1.split(',')
        if x in flight_dict:
            if y in flight_dict[x]:
                continue
            else:
                z=flight_dict.get(x)
                new=[]
                if type(z)==list:
                    for xx in z:
                        new.append(xx)
                else:
                    new.append(z)
                new.append(y)
                flight_dict[x]=new
        else:
            flight_dict[x]=[y]

for key in flight_dict:
    y=flight_dict.get(key)
    count=0
    for key1 in y:
        count+=1
    flight_dict[key]=count
    
#saving the output    
with open('task1.txt', 'w') as file:
     file.write(json.dumps(flight_dict))            
flight_dict         
```

```python
results1 = []

#mapper
for line in open(passenger):
    element1=line.strip()
    element1=line.split(',')
    new1=('%s,%s'%(element1[0],1))
    results1.append(new1)
    
results1
```

```python
#sorting the list
words1=list()

# Input from results
for line in results1:
    # Split the line into records - list of columns 
    word1=line.strip().split()
    # Append to list words
    words1.append(word1)
    
#sort list    
words1.sort()

#output
for line in words1:
    print(line)   
```

```python
#reducer
flight_dict1={}
for line in words1:
    for line1 in line:
        a,b=line1.split(',')
        if a in flight_dict1:
            value1=flight_dict1.get(a)
            value1=int(value1)
            value1+=1
            flight_dict1[a]=value1
        else:
            flight_dict1[a]=b
flight_dict1
```

```python
#finding the maximum key in the dictionary
max_key=max(flight_dict1,key=flight_dict1.get)
#finding the value of the key
max_value=flight_dict1.get(max_key)
#output
print("The Customer with Highest Number of Flight is: \n",max_key, " \n Number of Flights: \n", max_value)
```
